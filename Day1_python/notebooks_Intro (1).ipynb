```python
students = []
count = int(input('Enter students:'))
for i in range(count):
    name = input("name:")
    roll = input("roll:")
    marks = input('Marks:').split(',')
    marks = [int(m.strip()) for m in marks]
    total =sum(marks)
    avg = total/len(marks)

    if avg >=85:
        grade = 'A'
    elif avg>=70:  
        grade = 'B'
    elif avg >=50:
        grade = 'C'
    else:
        grade = 'Fail'
    students.append((name,roll,marks,total,avg,grade))  
for s in students:
    print(s)
```

```python
file = open('marks.txt', 'r')
marks = file.readlines()
file.close()

mark = []
for m in marks:
    mark.append(int(m.strip()))

avg = sum(mark)/len(mark)
passed = 0
failed = 0
for i in mark:
    if i >= 50:
        passed += 1
    else:
        failed +=1
report = open('reportcard.txt','w')
report.write(f"--- Student Report Card ---\n")
report.write(f"Average Marks: {avg:.2f}\n")
report.write(f"Students Passed: {passed}\n")
report.write(f"Students Failed: {failed}\n")
report.close()

print("Report generated successfully in reportcard.txt")
```
